```python
driver.get(url)
html = driver.page_source
print(html) 
```

```python
rest_list = driver.find_element(By.XPATH, "//div[@data-test-target='restaurants-list']")
divs = rest_list.find_elements(By.XPATH, "//div[contains(@data-test,'list_item')]") # "//p[contains(@class,'myclass')]"
```

```python
import re
divs = divs[1:6] # we take the only first 5, the 0 is usually ad

for div in divs:
    a_elem = div.find_element(By.TAG_NAME, "a")
    print(a_elem.get_attribute("href"))
    match = re.search(r'Reviews-(.+?).html', a_elem.get_attribute("href"))
    if match:
        desired_string = match.group(1)
        print(desired_string.replace('_', ' ').replace('-', ' '))
```
